```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df[df.time.str.isnumeric() == True]
```

```python
df['time'] = pd.to_datetime(df['time'], unit='s')
df['time'] = pd.DatetimeIndex(pd.date_range('2016-01-01 05:00', periods=len(df),  freq='min'))
df = df.set_index('time')
```

```python
df.columns = [i.replace(' [kW]', '') for i in df.columns]
df['Furnace'] = df[['Furnace 1','Furnace 2']].sum(axis=1)
df['Kitchen'] = df[['Kitchen 12','Kitchen 14','Kitchen 38']].sum(axis=1)
```

```python
df.drop(['Furnace 1','Furnace 2','Kitchen 12','Kitchen 14','Kitchen 38','icon','summary'], axis=1, inplace=True)
```

```python
fig = plt.subplots(figsize=(10, 8))
sns.heatmap(df[df.columns[0:15].tolist()].corr(), annot=True, fmt='.2f', vmin=-1.0, vmax=1.0, center=0)
plt.title('Energy Correlation', fontsize=12);
```

```python
df['Tdiff'] = df['apparentTemperature']-df['temperature']
```

```python
df.columns
```

```python
df['month'] = df.index.month
df['day'] = df.index.day
df['weekday'] = df.index.day_name()
df['hour'] = df.index.hour
df['minute'] = df.index.minute
```

```python
df['Sum_usage'] = df[list(df.columns[3:11])].sum(axis=1)+df[list(df.columns[24:25])].sum(axis=1)
```

```python
df['Usage_diff'] = df['House overall'] - df['Sum_usage']
```

```python
data = df[['House overall', 'Dishwasher', 'Home office', 'Fridge', 'Wine cellar', 'Garage door', 'Barn', 'Well', 'Microwave',
   'Living room', 'temperature', 'humidity', 'visibility','apparentTemperature', 'pressure', 'windSpeed', 'cloudCover',
   'windBearing', 'precipIntensity', 'dewPoint', 'precipProbability','Furnace','Kitchen', 'Tdiff', 'month', 'day', 'weekday', 'hour',
       'minute', 'Sum_usage', 'Usage_diff']]
```

```python
df['cloudCover'].replace(['cloudCover'], method='bfill', inplace=True)
df['cloudCover'] = df['cloudCover'].astype('float')
```
